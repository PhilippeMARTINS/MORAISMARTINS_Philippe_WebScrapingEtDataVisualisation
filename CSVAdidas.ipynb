```python
dff = dff.drop(columns=['Unnamed: 0'], axis=1)
dff = dff.drop(columns=['Unnamed: 0.1'], axis=1)
dff
```

```python
dff = dff.rename(columns={"1": "Modèle", "2": "Gamme", "3": "Prix", "4": "Img"})
dff
```

```python
dff['Prix'] = dff['Prix'].str.rstrip(' €')
dff
```
