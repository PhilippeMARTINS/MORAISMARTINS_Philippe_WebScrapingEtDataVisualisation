```python
df = pd.read_csv('AllReebokSneakers.csv')
df
```

```python
df = df.drop(columns=['Unnamed: 0'], axis=1)
df = df.drop(columns=['Unnamed: 0.1'], axis=1)
df
```

```python
df = df.rename(columns={"1": "Modèle", "2": "Gamme", "3": "Prix", "4": "Img"})
df
```

```python
Rds = df[ df['Prix'] == 'Rupture de stock' ].index
df.drop(Rds , inplace=True)
df
```

```python
neg = df[df['Prix'].str[0].isin(['-'])].index
df.drop(neg , inplace=True)
df
```

```python
df["Img"] = df["Img"].str[2:]
df
```

```python
df.to_csv("FinalReebokSneakers.csv")
```
